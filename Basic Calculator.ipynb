```python
def addition():
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    ans = num1 + num2
    
    count = 2
    
    continue_cal = 'y'
    
    while continue_cal.lower() == 'y':
        continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() not in ['y','n']:
            print("Please enter either y or n")
            continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() == 'y':
            num = float(input("Please enter the number: "))
            ans = ans + num
            count += 1
        else:
            break
            
        print(f"Current sum: {ans}")
    return(ans,count)
        

def substraction():
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    ans = num1 - num2
    
    count = 2
    
    continue_cal = 'y'
    
    while continue_cal.lower() == 'y':
        continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() not in ['y','n']:
            print("Please enter either y or n")
            continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() == 'y':
            num = float(input("Please enter the number: "))
            ans = ans - num
            count += 1
        else:
            break
            
        print(f"Current left: {ans}")
    return(ans,count)

def multiplication():
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    ans = num1 * num2
    
    count = 2
    
    continue_cal = 'y'
    
    while continue_cal.lower() == 'y':
        continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() not in ['y','n']:
            print("Please enter either y or n")
            continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() == 'y':
            num = float(input("Please enter the number: "))
            ans = ans * num
            count += 1
        else:
            break
            
        print(f"Current Product: {ans}")
    return(ans,count)

def division():
    
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    ans = num1 / num2
    
    count = 2
    
    continue_cal = 'y'
    
    while continue_cal.lower() == 'y':
        continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() not in ['y','n']:
            print("Please enter either y or n")
            continue_cal = input("Do you want to add more y/n: ")
        if continue_cal.lower() == 'y':
            num = float(input("Please enter the number: "))
            ans = ans / num
            count += 1
        else:
            break
            
        print(f"Current division: {ans}")
    return(ans,count)


def calc():
    option = False
    while option == False:
        
        print("""
        Welcome to simple calculator!
        For addition, press a:
        For substraction, press s:
        For multiplication, press m:
        For dicision, press d:
        To quit, press q:
        """)

        choice = input("Enter your choice: ")

        if choice == 'q':
            option = True
            continue
        
        elif choice == 'a':
            resp = addition()
            print(f"Final sum is: {resp[0]}, Data entered is: {resp[1]}")
        
        elif choice == 's':
            resp = substraction()
            print(f"Final left is: {resp[0]}, Data entered is: {resp[1]}")
        
        elif choice == 'm':
            resp = multiplication()
            print(f"Final product is: {resp[0]}, Data entered is: {resp[1]}")
        
        elif choice == 'd':
            resp = division()
            print(f"Final division is: {resp[0]}, Data entered is: {resp[1]}")
        
        else:
            print("Invalid input!!")

if __name__ == "__main__":
    calc()
```
